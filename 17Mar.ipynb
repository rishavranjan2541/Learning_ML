```python
# Analysis of a cricket 
over = [0,4,6,7,8,9]
team1 = [23,54,75,87,145,150]
team2 = [34,65,87,98,150,170]
plt.plot(over,team1, label = 'TEam1', marker ='d')
plt.plot(over, team2, label= 'Tema2', marker='v')
plt.title('Analysis of cricket team')
plt.xlabel('overs')
plt.ylabel('runs')
plt.legend()
```

```python
# SUBPLOT
import numpy as np
fig = plt.figure()

x1 = np.random.randint(0, 10, 5)
x2 = np.random.randint(0, 10,5)
y1 = np.random.randint(0, 10, 5)
y2 = np.random.randint(0, 10, 5)

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(224)

ax1.plot(x1,y1)
ax2.plot(x2,y2)
```

```python
# SCATTER PLOT(POINTS ARE SCATTERED)
a= [1,2,3,4,5,]
b= [2,3,4,5,6]
plt.title('SCATTER PLOT')
plt.scatter(a,b,)
plt.plot(a,b,c='r') # c here for the color and r is red color
```

```python
a = [1,2,3,4,5,6,7,8]
b = [5,4,6,2,3,1,7,5]
plt.scatter(a,b,color='lime', marker='*')
```

```python
# Line graph 
a = [1,2,3,4,5]
b = (2,6,1,3,5) # tuple
c = [9,8,7,3,4]
plt.plot(a,b,c='lime', label = 'line1', linestyle = 'dotted')
plt.plot(a,c, c = 'pink', label = 'line2', linestyle = 'dashed')
plt.legend() # to print the labels
```
