**BAR GRAPH**

```python
# bar graph
name = ['A', 'B','C','D','E']
num = [4,2,7,5,1]
plt.barh(name,num, color =['r','orange','k','pink','m'])
```

```python
x = np.array(['A', 'B','C','D'])
y = np.array([3,8,1,10])
plt.bar(x,y, color= 'orange', width = 0.3)
```

```python
# Pie Chart
parts = [1,2,3]
city = ['Banglore','mumbai', 'Delhi']
plt.pie(parts,labels = city, explode=[0.1,1,0])
```

```python
y = np.array([40,30,20,10])
brand = ['Xiomi','One Plus','Samsung','Apple']
plt.pie(y,labels=brand, explode = [0.1,0,0,1])
```
