```python
import numpy as np
import matplotlib.pyplot as plt
experience = np.arange(1,11)# print numbers from 1 to 10
np.random.seed(35)#In python , random values change everytime when cell is re run
#Hence , to keep the values constant , we use random.seed
salary = np.random.randint(20000,100000,10)
salary = np.sort(salary) # It sorts all the values in ascending order
print(experience)
print(salary)
```

```python
#divide data into i/p and o/p
x = experience.reshape(-1,1)
x
```

```python
y = salary
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y) # mapping the values of x and y in linear regression
```

```python
# predict variable
y_pred = model. predict(x)
y_pred # predicted values
```

```python
y # actual values
```

```python
# Individual prediction
model.predict([[2.5]])
```

```python
# cross verify
m = model.coef_
m
```

```python
c = model.intercept_
c
```

```python
# y - mx + c
m*2.5 + c
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/Social_Network_Ads.csv')
df
```

```python
df.info()
```

```python
# Purchased colum : 0 and 1
# 0 - not purchased 
# 1 - purchased
# we are not considering user id and the gender column
# purchase column  n- output
# age and estimated salary is input
```

```python
# divide data into  i/p and o/p
x = df.iloc[:, 2:4].values
x
```

```python
y = df.iloc[:,4].values
y
```

```python
df['Purchased'].value_counts()
```

```python
# train and test variables 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state = 0)
```

```python
x.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
# apply classifier / regressor/clusterer
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# model fitting
model.fit(x_train, y_train)
```

```python
# predictor variable
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
# accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```

```python
# Individual prediction 
model.predict([[25,250000]])
```
