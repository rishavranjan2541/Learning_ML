```python
#1. READ THE IMAGE
import cv2 # importing the ipencv library
img = cv2.imread('abc.jpeg')  # reading our image
cv2.imshow('OUTPUT',img)  # It will display the image
cv2.waitKey(0) # waitKey(0) - tje image will stay forever until closed
cv2.destroyAllWindows()  #it closes all the python output windows if open
```

```python
#2. CREATE A DUPLICATE IMAGE
import cv2
img = cv2.imread('abc.jpeg')
cv2.imshow('ORIGINAL IMAGE',img)
cv2.imshow('acer.png',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 3. READ THE INFO ABOUT THE IMAGE
import cv2
img = cv2.imread('abc.jpeg') # read the image
print(img.shape)


# OUTPUT 
# (192, 262, 3)
#  192 - Height of image in pixels
#  262 - Width of image in pixels
#  3   - channel value/depth(RGB)
```
